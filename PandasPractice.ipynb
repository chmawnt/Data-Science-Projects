```python
import numpy as np
import pandas as pd

df =pd.read_csv('/content/sample_data/Salaries.csv')
#df.info()
df.head()
```

```python
# What is the average BasePay ?**
average_basepay = df['BasePay'].mean()
print(average_basepay)
# ** What is the highest amount of OvertimePay in the dataset ? **
highest_overtimepay = df['OvertimePay'].max()
print(highest_overtimepay)
```

```python
#What is the job title of  JOSEPH DRISCOLL ?
#Note: Use all caps, otherwise you may get an answer that doesn't match up
 # (there is also a lowercase Joseph Driscoll). **

print(df[df['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle'])

```

```python
# How much does JOSEPH DRISCOLL make (including benefits)? **
print(df[df['EmployeeName']=='JOSEPH DRISCOLL']['TotalPayBenefits'])
```

```python
# What is the name of highest paid person (including benefits)?**
df[df['TotalPayBenefits']==df['TotalPayBenefits'].max()]
#df.loc[df['TotalPayBenefits'].idxmax()]
#df[df['TotalPayBenefits']==df['TotalPayBenefits'].max()]
```

```python
# What is the name of lowest paid person (including benefits)?
 # Do you notice something strange about how much he or she is paid?**
df[df['TotalPayBenefits']==df['TotalPayBenefits'].min()]
#df.loc[df['TotalPayBenefits'].idxmin()]
```

```python
#What was the average (mean) BasePay of all employees per year? (2011-2014) ? **
df.groupby('Year')['BasePay'].mean()
```

```python
# How many unique job titles are there? **
df['JobTitle'].nunique()

#** What are the top 5 most common jobs? **
df['JobTitle'].value_counts().head(5)

#** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **
sum(df[df['Year']==2013]['JobTitle'].value_counts()==1)

# ** How many people have the word Chief in their job title? (This is pretty tricky) **
#sal[sal["JobTitle"].apply(lambda x : "chief" in x.lower())]["JobTitle"].count()
import re
jobtitles = list(df['JobTitle'])
count = 0

for job in jobtitles:
    re.search('chief*', job.lower())
    count += 1

print(count)

def chief_string(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False

sum(df['JobTitle'].apply(lambda x: chief_string(x)))
```

```python
# Bonus: Is there a correlation between length of the Job Title string and Salary? **
df['title_len'] = df['JobTitle'].apply(len)

df[['title_len','TotalPayBenefits']].corr()
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/sample_data/Ecommerce Purchases')
df.head()
```

```python
#** How many rows and columns are there? **
df.info()
df['Purchase Price'].mean()
df['Purchase Price'].max()
df['Purchase Price'].min()
```

```python
#** How many people have English 'en' as their Language of choice on the website? **
df[df['Language']=='en'].count()
```

```python
# How many people have the job title of "Lawyer" ? **
df[df['Job']=='Lawyer'].info()
```

```python
#** How many people made the purchase during the AM and how many people made the purchase during PM ? **
#hint: Check out [value_counts()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html) ) **
df['AM or PM'].value_counts()
```

```python
#** What are the 5 most common Job Titles? **
df['Job'].value_counts().head(5)
```

```python
#** Someone made a purchase that came from
# Lot: "90 WT" , what was the Purchase Price for this transaction? **
df[df['Lot']=="90 WT"]['Purchase Price']
```

```python
# What is the email of the person with the following Credit Card Number: 4926535242672853 **
df[df['Credit Card']==4926535242672853]['Email']
```

```python
#How many people have American Express as their Credit Card Provider
# *and* made a purchase above $95 ?**
df[(df['CC Provider']=='American Express') & (df['Purchase Price']>95)].count()

```

```python
#** Hard: How many people have a credit card that expires in 2025? **

df[df['CC Exp Date'].apply(lambda exp: exp[3:]=='25')].count()
```

```python
# What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...)
df['Email'].apply(lambda email:email.split('@')[1]).value_counts()
```
