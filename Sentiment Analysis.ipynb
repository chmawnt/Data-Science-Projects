```python
from textblob import TextBlob
import pandas as pd
```

```python
def sentiment_analysis(text):
  text = text.lower()
  blob_text = TextBlob(text)
  sentiment = blob_text.sentiment.polarity

  if sentiment >0:
    return 'Positive'
  elif sentiment <0:
    return 'Negative'
  else:
    return 'Neutral'
```

```python
df = pd.read_csv('/content/drive/MyDrive/redmi6.csv', encoding='ISO-8859-1')
df = df[['Customer name','Comments']]
df.head()
```

```python
df['label'] = df['Comments'].apply(sentiment_analysis)
df
```
